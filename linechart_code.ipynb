```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Dades extretes del github de Our World in Data: https://github.com/owid/energy-data

Article de diari sobre la visualitzaci√≥ de dades: https://www.newyorker.com/magazine/2021/06/21/when-graphs-are-a-matter-of-life-and-death

```python
data = pd.read_csv("linechart_data.csv")

data = data.query("country == 'Spain' and year > 2009")
data = data[["year", "fossil_electricity", "renewables_electricity"]]

f = plt.figure()
f.set_figheight(8)
f.set_figwidth(10)
sns.lineplot(data=data, x="year", y="fossil_electricity",
             label="F√≤ssil", color="brown")
sns.lineplot(data=data, x="year", y="renewables_electricity",
             label="Renovable", color="lightgreen")
plt.legend()
plt.title("Generaci√≥ d'electricitat a Espanya per tipus de font")
plt.xlabel("Any")
plt.ylabel("Generaci√≥ (TWh)")
plt.savefig("linechart.png")
```
